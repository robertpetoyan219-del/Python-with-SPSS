```python
# pip install deep-translator openpyxl pandas
import pandas as pd
from openpyxl import load_workbook
from deep_translator import GoogleTranslator
import time

INPUT_FILE = "Converse_Competitive_Desk_Research_EN.xlsx"
OUTPUT_FILE = "your_file_armenian.xlsx"

translator = GoogleTranslator(source='auto', target='hy')  # 'hy' = Armenian

def translate_cell(text):
    if not isinstance(text, str) or not text.strip():
        return text
    try:
        time.sleep(0.1)  # Avoid rate limiting
        return translator.translate(text)
    except Exception as e:
        print(f"Failed: {text[:30]}... → {e}")
        return text  # Keep original on failure

wb = load_workbook(INPUT_FILE)

for sheet_name in wb.sheetnames:
    ws = wb[sheet_name]
    print(f"Processing sheet: {sheet_name}")
    for row in ws.iter_rows():
        for cell in row:
            if isinstance(cell.value, str) and cell.value.strip():
                cell.value = translate_cell(cell.value)

wb.save(OUTPUT_FILE)
print(f"Saved: {OUTPUT_FILE}")
```
