## How to easily work if no SPSS ##

```python
# 1. See ALL column names
print(df.columns.tolist())  # List of all column names

# 2. See column names WITH their question labels (this is key!)
for col in df.columns:
    label = meta.column_names_to_labels.get(col, 'No label')
    print(f"{col}: {label}")

# 3. Export to Excel to browse easily
pd.DataFrame({
    'Variable': df.columns,
    'Label': [meta.column_names_to_labels.get(col, '') for col in df.columns]
}).to_excel('column_guide.xlsx', index=False)
```

## Data Analysis with Pandas

```python
import pandas as pd

# Basic exploration
df.shape  # (rows, columns)
df.info()  # Data types, non-null counts
df.describe()  # Summary statistics for numeric columns

# Single column analysis
df['Q1'].value_counts()  # Frequency table
df['Q1'].value_counts(normalize=True)  # Percentages
df['age'].mean()  # Average
df['age'].median()  # Median

# Cross-tabulation (relationship between 2 variables)
pd.crosstab(df['gender'], df['party_preference'])
pd.crosstab(df['gender'], df['party_preference'], normalize='index')  # Row %

# Filter and analyze
young_voters = df[df['age'] < 30]
young_voters['party_preference'].value_counts()

# Group analysis
df.groupby('region')['satisfaction'].mean()
df.groupby(['region', 'gender'])['satisfaction'].agg(['mean', 'count', 'std'])

# Correlation between numeric variables
df[['age', 'income', 'satisfaction']].corr()
```

## Data Analysis with Pyreadstat

```python
import pyreadstat

df, meta = pyreadstat.read_sav('your_file.sav')

# The KEY advantage: value labels (e.g., 1='Male', 2='Female')
print(meta.variable_value_labels)

# Example: Q1 might have codes like 1, 2, 3
# But they mean 'Strongly Agree', 'Agree', 'Disagree'

# Get value labels for specific variable
q1_labels = meta.variable_value_labels.get('Q1', {})
print(q1_labels)  # {1: 'Strongly Agree', 2: 'Agree', ...}

# Replace codes with labels
df['Q1_labeled'] = df['Q1'].map(q1_labels)
df['Q1_labeled'].value_counts()  # Now shows actual meanings!

# Or apply to entire dataframe
for col in df.columns:
    if col in meta.variable_value_labels:
        df[f'{col}_labeled'] = df[col].map(meta.variable_value_labels[col])

# Get question text
question_text = meta.column_names_to_labels.get('Q15')
print(f"Q15 asks: {question_text}")
```

## Combined approach ##

```python
import pandas as pd
import pyreadstat

df, meta = pyreadstat.read_sav('your_file.sav')

# Create a readable version with labels
df_readable = df.copy()
for col in df.columns:
    if col in meta.variable_value_labels:
        df_readable[col] = df[col].map(meta.variable_value_labels[col])

# Now analyze the readable version
df_readable['party_preference'].value_counts()
# Shows: "Conservative: 150, Labour: 200..." 
# Instead of: "1: 150, 2: 200..."

# Cross-tab with actual labels
pd.crosstab(df_readable['gender'], df_readable['voting_intention'])
```
